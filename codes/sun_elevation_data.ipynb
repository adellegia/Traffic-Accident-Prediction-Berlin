```python
tzone = tz.gettz('Europe/Berlin')
```

```python
# EXAMPLE:
latitude = 52.509
longitude = 13.385

date = datetime.datetime(2022, 4, 1, 16, 0, 0, tzinfo=tzone)
sun_altitude = round(get_altitude(latitude, longitude, date),3)
sun_altitude
```

```python
def get_sun_altitude(year,month,day,hour,lat,long):
  date = datetime.datetime(year, month, day, hour, 0, 0, tzinfo=tzone)
  sun_altitude = round(get_altitude(lat, long, date),3)
  return sun_altitude
```
